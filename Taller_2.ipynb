Si el empleado es mayor de 55 años disfrutará de un bono de prepensión correspondiente al 5% de su sueldo básico.

```python
edad = 59
sueldo_basico = 3000000
if edad > 55:
  bono = sueldo_basico * 0.05
  total = sueldo_basico + bono
  print("Recibe bono de prepension de =${:,.0f} y el total a pagar es= {:,.0f}".format(bono, total))
else:
  print("Unicamente sueldo")
```

Si el sueldo básico está entre 1000000 y 1500000 tendrá una comisión del 2% sobre el valor del sueldo; Si el sueldo básico está entre 1500001 y 2000000 tendrá una comisión del 5% sobre el valor del sueldo; para todos los demás casos no habrá comisión.

```python
sueldo_basico = 3000000
com1 = sueldo_basico * 0.02
com2 = sueldo_basico * 0.05
total1 = sueldo_basico + com1
total2 = sueldo_basico + com2
if sueldo_basico >= 1000000 and sueldo_basico <= 1500000:
  print("Recibe comision de =${:,.0f} y el total a pagar es=${:,.0f}".format(com1, total1))
elif sueldo_basico >= 1500001 and sueldo_basico <= 2000000:
  print("Recibe comision de =${:,.0f} y el total a pagar es=${:,.0f}".format(com2, total2))
else:
  print("no aplica comision")
```
